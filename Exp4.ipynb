```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
x_train=x_train.reshape(-1,784).astype("float32")/255
x_test  = x_test.reshape(-1, 784).astype("float32") / 255
```

```python
def create_model(learning_rate=0.01,hidden_units=128):
    model = keras.Sequential([
    layers.Dense(hidden_units,activation='relu',input_shape=(784,)),
    layers.Dense(64,activation='relu'),
    layers.Dense(10,activation='softmax')
    ])

    optimizer=optimizers.SGD(learning_rate=learning_rate,momentum=0.9)
    model.compile(optimizer=optimizer,
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model

```

```python
print("==== Manual Hyperparameter Tuning ====")
```

```python
learning_rates = [0.01, 0.001]
hidden_units_list = [64, 128]
batch_sizes = [32, 64]
```

```python
for lr in learning_rates:
    for hu in hidden_units_list:
        for bs in batch_sizes:
            print(f"\nğŸ”¹ Training with lr={lr}, hidden_units={hu}, batch_size={bs}")
            model = create_model(learning_rate=lr, hidden_units=hu)
            history = model.fit(
                x_train, y_train,
                epochs=5, batch_size=bs,
                validation_data=(x_test, y_test),
                verbose=0
            )
            test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
            print(f"âœ… Test Accuracy = {test_acc:.4f}")
```

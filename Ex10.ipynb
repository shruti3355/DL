```python
from transformers import pipeline
import gradio as gr
```

```python
translator = pipeline("translation_en_to_hi", model="Helsinki-NLP/opus-mt-en-hi")
```

```python
def translate_text(text):
    if text.strip()=="":
      return "Please enter some text to translate"
    result=translator(text)
    return result[0]['translation_text']
```

```python
interface = gr.Interface(
    fn=translate_text,
    inputs=gr.Textbox(lines=3, placeholder="Enter English text here..."),
    outputs="text",
    title="English to Hindi Translator using Pre-trained LLM",
    description="This app translates English text into Hindi using a pre-trained large language model from Hugging Face."
)
```

```python
interface.launch()
```
